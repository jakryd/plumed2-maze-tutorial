```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
X = np.loadtxt('colvar.dat')
```

```python
fig, ax = plt.subplots(1, 1)
ax.plot(X[:,0], X[:,1])
ax.set_xlabel('Time [ps]')
ax.set_ylabel('Distance [nm]')
fig.savefig('time-dist.png')
```

```python
fig, ax = plt.subplots(1, 1)
ax.plot(X[:,0], X[:,5])
ax.set_xlabel('Time [ps]')
ax.set_ylabel('Loss')
fig.savefig('time-loss.png')
```

```python
fig, ax = plt.subplots(1, 1)
ax.plot(X[:,0], X[:,6])
ax.set_xlabel('Time [ps]')
ax.set_ylabel('Radius [nm]')
fig.savefig('time-radius.png')
```

```python
fig, ax = plt.subplots(1, 1)
ax.plot(X[:,0], X[:,7])
ax.set_xlabel('Time [ps]')
ax.set_ylabel('Bias [kJ/mol]')
fig.savefig('time-bias.png')
```

```python
fig, ax = plt.subplots(1, 1)
ax.hist(X[:,7], bins=100)
ax.set_xlabel('Time [ps]')
ax.set_ylabel('Bias [kJ/mol]')
#fig.savefig('time-bias.png')
```
